### Covariance:

* Cov tells us if the variables are positively or inversely related (basically direction of the correlation)
* If we select 2 random variables then It tells the relationship btw them and at what degree/extend both the variables tend to change together  

* Cov just gives the direction of the relationship between the variables

$$
\text{cov}(x, y) = \frac{\sum_{i=1}^{N} (x_i - \bar{x})(y_i - \bar{y})}{N-1}
$$

* cov_{x,y}	=	covariance between variable x and y

* $ x_{i} $	=	data value of x

* $ y_{i} $	=	data value of y

* $ \bar{x} $	=	mean of x

* $ \bar{y} $	=	mean of y
* N	=	number of data values
* (N-1) for sample data 
* N for population data

### Correlation

* Here corr tells the magnitude to the value of Cov. basically to give the scale of the value of Cov Correlation value is used. So hence Corr is measured btw -1 to 1

* Corr gives both the direction and scale of the relationship between the variables that is whether it is this value and direction like positive or negatively

* It is also called Correlation Coefficient Or Pearson's Correlation Coefficient (r)

$$
r = \frac{\sum_{i=1}^{N} (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{N} (x_i - \bar{x})^2} \sqrt{\sum_{i=1}^{N} (y_i - \bar{y})^2}}
$$

$$
r = \frac{\text{cov}(x, y)}{\sigma_x \sigma_y}
$$

* cov_{x,y}	=	covariance between variable x and y

* $ x_{i} $	=	data value of x

* $ y_{i} $	=	data value of y

* $ \bar{x} $	=	mean of x

* $ \bar{y} $	=	mean of y
* N	=	number of data values
* (N-1) for sample data 
* N for population data

### Step-by-Step Explanation
1. **Calculate Means**:
   $
   \bar{x} = \frac{2 + 4 + 6 + 8}{4} = 5, \quad \bar{y} = \frac{5 + 8 + 11 + 14}{4} = 9.5
   $
---
2. **Compute Covariance**:
   $
   \text{cov}(x, y) = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{N - 1}
   $
---
---
   Substituting values:
   $
   \text{cov}(x, y) = \frac{(2 - 5)(5 - 9.5) + (4 - 5)(8 - 9.5) + (6 - 5)(11 - 9.5) + (8 - 5)(14 - 9.5)}{3}
   $
---  
---
   Simplify:
   $
   \text{cov}(x, y) = \frac{(-3)(-4.5) + (-1)(-1.5) + (1)(1.5) + (3)(4.5)}{3} = \frac{13.5 + 1.5 + 1.5 + 13.5}{3} = 10.0
   $
---
---
3. **Calculate Standard Deviations**:
---
   For \(x\):
   $
   \sigma_x = \sqrt{\frac{\sum (x_i - \bar{x})^2}{N - 1}}
   $
---
   Substituting:
   
---
   $
   \sigma_x = \sqrt{\frac{(-3)^2 + (-1)^2 + (1)^2 + (3)^2}{3}} = \sqrt{\frac{9 + 1 + 1 + 9}{3}} = \sqrt{6.67} \approx 2.58
   $
   Similarly, compute for \(y\):
   $
---
   \sigma_y = \sqrt{\frac{(-4.5)^2 + (-1.5)^2 + (1.5)^2 + (4.5)^2}{3}} = \sqrt{\frac{20.25 + 2.25 + 2.25 + 20.25}{3}} = \sqrt{15.0} \approx 3.87
   $

4. **Compute Correlation Coefficient**:
---
   $
   r = \frac{\text{cov}(x, y)}{\sigma_x \sigma_y}
   $
---
   Substituting:
   
---
   $
   r = \frac{10.0}{2.58 \times 3.87} \approx \frac{10.0}{10.0} = 1.0

### Structure of a Covariance Matrix
For a dataset with \(n\) variables (\(X_1, X_2, \dots, X_n\)), the covariance matrix is an \(n \times n\) matrix where:

- Each element \(C_{ij}\) represents the covariance between the \(i\)-th and \(j\)-th variables.
- The diagonal elements (\(C_{ii}\)) represent the variances of the variables.
- The off-diagonal elements (\(C_{ij}\)) represent the covariances between the variables.

$
C = 
\begin{bmatrix}
\text{Var}(X_1) & \text{Cov}(X_1, X_2) & \dots & \text{Cov}(X_1, X_n) \\
\text{Cov}(X_2, X_1) & \text{Var}(X_2) & \dots & \text{Cov}(X_2, X_n) \\
\vdots & \vdots & \ddots & \vdots \\
\text{Cov}(X_n, X_1) & \text{Cov}(X_n, X_2) & \dots & \text{Var}(X_n)
\end{bmatrix}
$

---

### Properties of a Covariance Matrix
1. **Symmetry**:
   $
   \text{Cov}(X_i, X_j) = \text{Cov}(X_j, X_i)
   $
   So, the matrix is symmetric.

2. **Diagonal Elements**:
   The diagonal elements represent the variances:
   $
   \text{Var}(X_i) = \text{Cov}(X_i, X_i)
   $

3. **Off-Diagonal Elements**:
   The off-diagonal elements represent the covariances:
   $
   \text{Cov}(X_i, X_j)
   $

---

### Example

Suppose we have a dataset with two variables \(X\) and \(Y\), with the following data points:

| \(X\) | \(Y\) |
|------|------|
| 2    | 5    |
| 4    | 8    |
| 6    | 11   |
| 8    | 14   |

#### Step 1: Compute Means
$
\bar{X} = 5, \quad \bar{Y} = 9.5
$

#### Step 2: Compute Variances and Covariance
$
\text{Var}(X) = \frac{\sum (X_i - \bar{X})^2}{N-1} = \frac{9 + 1 + 1 + 9}{3} = 6.67
$
$
\text{Var}(Y) = \frac{\sum (Y_i - \bar{Y})^2}{N-1} = \frac{20.25 + 2.25 + 2.25 + 20.25}{3} = 15.0
$
$
\text{Cov}(X, Y) = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{N-1} = 10.0
$

#### Step 3: Form the Covariance Matrix
$
C = 
\begin{bmatrix}
\text{Var}(X) & \text{Cov}(X, Y) \\
\text{Cov}(Y, X) & \text{Var}(Y)
\end{bmatrix}
=
\begin{bmatrix}
6.67 & 10.0 \\
10.0 & 15.0
\end{bmatrix}
$

---


### Interpretation
- **Diagonal Elements**:
  - \(6.67\): Variance of \(X\).
  - \(15.0\): Variance of \(Y\).
- **Off-Diagonal Elements**:
  - \(10.0\): Covariance between \(X\) and \(Y\).

The covariance matrix gives a summary of how the variables vary individually and in relation to each other.


$
C = 
\begin{pmatrix}
\sigma(x, x) & \sigma(x, y) \\
\sigma(y, x) & \sigma(y, y)
\end{pmatrix}
$


---

- $ \sigma(x, x) $ represents the variance of $ x $ ($ \text{Var}(x) $).
- $ \sigma(y, y) $ represents the variance of $ y $ ($ \text{Var}(y) $).
- $ \sigma(x, y) = \sigma(y, x) $ represents the covariance between $ x $ and $ y $.

This notation is frequently used in mathematical texts and conveys the same information as the standard covariance matrix but replaces $ \text{Cov}(X, Y) $ with $ \sigma(X, Y) $ for symmetry and compactness.



---

### Clarification of Symmetry
Because covariance is symmetric, $ \sigma(x, y) $ = $ \sigma(y, x) $, this covariance matrix is also symmetric.

---

### Basics to Understand Eigenvalues and Eigenvectors  

To grasp **eigenvalues** and **eigenvectors**, it's helpful to first review a few foundational concepts in linear algebra.

---

### **Vectors and Matrices**
- A **vector** is a quantity with both magnitude and direction. Example: $v = \begin{bmatrix} 3 \\ 2 \end{bmatrix} $ is a 2D vector.  
- A **matrix** represents a transformation (like rotation, scaling, or reflection) that can be applied to vectors. Example:  
  $
  A = \begin{bmatrix} 2 & 1 \\ 1 & 3 \end{bmatrix}
  $
  Applying this transformation to a vector $v $ gives:
  $
  A v = \text{new vector.}
  $

---

### **Linear Transformation and Its Effect**
When a matrix $A $ acts on a vector $v $:
- The vector might **rotate**, **stretch**, or **shrink**.  
- **Eigenvectors** are special because they don’t change direction during the transformation; they only get scaled (stretched or shrunk).  
  - Example: A vector pointing in a specific direction remains aligned in that direction.

---

###  **Definition of Eigenvalues and Eigenvectors**
### Definitions

1. **Eigenvalues ($\lambda $):**
   An eigenvalue is a scalar that indicates how much an eigenvector is scaled during a linear transformation. 

2. **Eigenvectors ($v $):**
   An eigenvector of a square matrix $A $ is a non-zero vector that only changes in scale when the matrix is applied to it.

The relationship between a matrix $A $, an eigenvector $v $, and its corresponding eigenvalue $\lambda $ is given by:

$
A v = \lambda v
$

Where:
- $A $ is an $n \times n $ matrix.
- $v $ is an $n $-dimensional vector (eigenvector).
- $\lambda $ is a scalar (eigenvalue).

---

### Characteristic Equation

To find the eigenvalues of $A $, solve the **characteristic equation**:

$
\det(A - \lambda I) = 0
$

Where:
- $I $ is the identity matrix of the same dimension as $A $.
- $\det $ denotes the determinant of the matrix.

---

---

###  **Key Characteristics**
1. **Eigenvectors**: Represent the "invariant directions" of the transformation.  
   - Example: If $A $ represents a rotation + scaling, eigenvectors are the directions that don’t rotate.  

2. **Eigenvalues**: Indicate the "scaling factor" along the eigenvector.  
   - Example: $\lambda > 1 $ means the eigenvector is stretched; $0 < \lambda < 1 $ means it’s shrunk.  

---

### **How to Find Eigenvalues and Eigenvectors**
To calculate eigenvalues and eigenvectors of $A $:  
1. **Eigenvalue ($\lambda $):** Solve the **characteristic equation**:  
   $
   \det(A - \lambda I) = 0
   $
   - Here, $I $ is the identity matrix, and $\det $ is the determinant.  

2. **Eigenvector ($v $):** Solve the equation $(A - \lambda I)v = 0 $ using each eigenvalue $\lambda $.  
3. **Always take larger eigen value**

---

### Example: Step-by-Step Calculation

#### Matrix $A $:
$
A = \begin{bmatrix} 4 & 1 \\ 2 & 3 \end{bmatrix}
$

1. **Find Eigenvalues ($\lambda $):**
   $
   \det(A - \lambda I) = \det\left(\begin{bmatrix} 4-\lambda & 1 \\ 2 & 3-\lambda \end{bmatrix}\right)
   $
   Compute the determinant:
   $
   \det(A - \lambda I) = (4-\lambda)(3-\lambda) - (2)(1) = \lambda^2 - 7\lambda + 10 = 0
   $
   Solve $\lambda^2 - 7\lambda + 10 = 0 $:
   $
   \lambda = 5, \; \lambda = 2
   $

2. **Find Eigenvectors:**
   For each $\lambda $, solve $(A - \lambda I)v = 0 $.

   - For $\lambda = 5 $:
     $
     A - 5I = \begin{bmatrix} -1 & 1 \\ 2 & -2 \end{bmatrix}
     $
     Solve $(A - 5I)v = 0 $:
     $
     v = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
     $

   - For $\lambda = 2 $:
     $
     A - 2I = \begin{bmatrix} 2 & 1 \\ 2 & 1 \end{bmatrix}
     $
     Solve $(A - 2I)v = 0 $:
     $
     v = \begin{bmatrix} -1 \\ 2 \end{bmatrix}
     $

---
###  **Visualization**
- The eigenvectors $v_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix} $ and $v_2 = \begin{bmatrix} -1 \\ 2 \end{bmatrix} $ are the special directions in the transformation.  
- The eigenvalues $\lambda_1 = 5 $ and $\lambda_2 = 2 $ show the scaling factors along these directions.

# Principal Component Analysis (PCA)

[Principal component analysis](https://en.wikipedia.org/wiki/Principal_component_analysis) (PCA) is a mechanism for reducing the number of dimensions in data. It enjoys a number of practical applications in machine learning, including obfuscating sensitive or proprietary data, eliminating multicollinearity, removing noise from datasets, and visualizing high-dimensional data by reducing it to two or three dimensions that can be plotted. PCA doesn't directly make machine-learning models more accurate, but it can help them train faster and generalize better. And PCA is easy to perform using Scikit's [PCA](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html) class. Let's use it to understand how PCA works and why it can be beneficial in machine learning.



````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
from sklearn.datasets import fetch_lfw_people

faces=fetch_lfw_people(min_faces_per_person=10)
#print(faces.target_names)
print(faces.images.shape)
````

````python
fig,ax=plt.subplots(nrows=3,ncols=8,figsize=(18,10))

for i, ax in enumerate(ax.flat):
    ax.imshow(faces.images[i],cmap='gist_gray')
    ax.set(xticks=[],yticks=[] , xlabel=faces.target_names[faces.target[i]])
````

Your understanding is close, but there’s an important distinction to clarify regarding the **number of features** and the **dimensionality reduction using PCA**.

---

### **1. Original Dataset Structure**
- So In the above dataset, each image is of size $ 62 \times 47 $ pixels.
- When flattened, each image becomes a **1D array with $62 \times 47 = 2914$ values (or features)**.
- Thus, the **number of features in the original dataset is $2914$.


````python
from sklearn.decomposition import PCA
````

````python
pca=PCA(n_components=150,random_state=0)
pca_faces=pca.fit_transform(faces.data)
unpca_faces=pca.inverse_transform(pca_faces).reshape(4324, 62, 47)
````

````python
cov_mat=np.cov(pca_faces.T)
cov_mat

print(cov_mat,' \n shape - ',cov_mat.shape)
````

````python
eig_val,eig_vec=np.linalg.eig(cov_mat)

print('eig_val - ',eig_val,' \n\n eig_vec - ',eig_vec)
````

````python
# Cumilative explained variance ratio

np.cumsum(pca.explained_variance_ratio_)
````

````python
fig,ax=plt.subplots(nrows=3,ncols=8,figsize=(18,10))

for i, ax in enumerate(ax.flat):
    ax.imshow(unpca_faces[i],cmap='gist_gray')
    ax.set(xticks=[],yticks=[] , xlabel=faces.target_names[faces.target[i]])
````

Compare the before and after images. Does it look as if 95% of the information was discarded? In fact, you can compute how much of the original information was retained by summing the explained variances. In this case, it's almost 95%:

````python
np.sum(pca.explained_variance_ratio_)
````

**pca.explained_variance_ratio_ is an array where each element represents the proportion of the total variance in the data explained by each principal component.**

#### Purpose:
The **total explained variance ratio** is useful to understand how much of the original data's variance is captured by the principal components you have kept. Ideally, it should be close to 1 (or 100%) if all principal components are included.


* **By examining the explained variance ratio, you can determine how much information is retained by the reduced set of principal components. If you choose fewer components, you lose some information, but the goal is to keep most of the important variance while reducing the number of dimensions (features)**

A logical question to ask is what is the "right" number of components? In other words, what number of components strikes the best balance between reducing the number of dimensions in the dataset and retaining most of the information? One way to answer that question is a scree plot, which plots the proportion of explained variance for each dimension. Here is a scree plot for the PCA transform used on the facial images.

````python
fig,ax1=plt.subplots()
ax1.plot(pca.explained_variance_ratio_)

ax1.set_xlabel('Principal Component')
ax1.set_ylabel('explained_variance');



fig,ax2=plt.subplots()
ax2.plot(np.cumsum(pca.explained_variance_ratio_))

ax2.set_xlabel('Principal Component')
ax2.set_ylabel('explained_variance')
ax2.set_title('Another way to look at it is to plot the sum of the variances as a function of component count.')
;
````
