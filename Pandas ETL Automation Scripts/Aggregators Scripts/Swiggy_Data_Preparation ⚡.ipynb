### Droping first row of table and renaming the headers (i.e top columns)


```python

db=pd.DataFrame(pivoted_data) ## creating new dataframe with pivoted data
db=db.reset_index()
#db.drop(columns=['Payout_period_no'],inplace=True)

db.fillna(0,inplace=True)
db.columns=db.columns.droplevel(0)
db.columns.values[0]='Serial_no' # renaming the colmns
db.columns.values[1]='Res_name' # renaming the colmns
db.columns.values[2]='Parameter' # renaming the colmns
```

### Adding Ad% values as rows


```python
s=db[db['Particular']=='HIGH_PRIORITY'].groupby(['Res_name'])[db.columns.values[3:]].sum()
#s

x=db[db['Particular']=='Total Customer Payable '].groupby(['Res_name'])[db.columns.values[3:]].sum()
#x


#(s/x)*100
Ad_percent=pd.DataFrame(round(((s/x)*100),0))
Ad_percent.insert(0,'Particular',value='Ad%',allow_duplicates=False)
Ad_percent=Ad_percent.reset_index() 
Ad_percent.fillna(0,inplace=True)
Ad_percent
```

### Adding Credit% values as rows


```python
p=db[db['Particular']=='Net Payout  (E - F - G)'].groupby(['Res_name'])[db.columns.values[3:]].sum()
#p

e=db[db['Particular']=='Total Customer Payable '].groupby(['Res_name'])[db.columns.values[3:]].sum()
#e


#(p/e)*100
Credit_percent=pd.DataFrame(round(((p/e)*100),0))
Credit_percent.insert(0,'Particular',value='Credit%',allow_duplicates=False)
Credit_percent=Credit_percent.reset_index() 
Credit_percent.fillna(0,inplace=True)
Credit_percent
```

### Adding Long distance charges% values as rows


```python
c=db[db['Particular']=='LONG_DISTANCE_PACK_CHARGES'].groupby(['Res_name'])[db.columns.values[3:]].sum()
#c

v=db[db['Particular']=='Total Customer Payable '].groupby(['Res_name'])[db.columns.values[3:]].sum()
#v


#(c/v)*100
LD_percent=pd.DataFrame(round(((c/v)*100),0))
LD_percent.insert(0,'Particular',value='Long_distance%',allow_duplicates=False)
LD_percent=LD_percent.reset_index() 
LD_percent.fillna(0,inplace=True)
LD_percent
```

### Adding payout% values as rows


```python
f=db[db['Particular']=='Net Payout  (E - F - G)'].groupby(['Res_name'])[db.columns.values[3:]].sum()
#f

q=db[db['Particular']=='Total Customer Payable '].groupby(['Res_name'])[db.columns.values[3:]].sum()
#q


#(c/v)*100
payout_percent=pd.DataFrame(round(((f/q)*100),1))
payout_percent.insert(0,'Particular',value='Payout%',allow_duplicates=False)
payout_percent=payout_percent.reset_index() 
payout_percent.fillna(0,inplace=True)
payout_percent
```

### Adding credit_per_order values as rows


```python
n=db[db['Particular']=='Number of orders'].groupby(['Res_name'])[db.columns.values[3:]].sum()
#n

j=db[db['Particular']=='Total Customer Payable '].groupby(['Res_name'])[db.columns.values[3:]].sum()
#j


#(c/v)*100
credit_per_order=pd.DataFrame(round((j/n),1))
credit_per_order.insert(0,'Particular',value='Credit_per_order',allow_duplicates=False)
credit_per_order=credit_per_order.reset_index() 
credit_per_order.fillna(0,inplace=True)
credit_per_order
```
